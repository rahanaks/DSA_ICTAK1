# Visualization 1: Boxplot

## Boxplot for SL

```python
sns.boxplot(x='Classification', y='SL', data = iris)
plt.title('Boxplot for SL', color='blue', size = 20)
plt.show()
```

## Boxplot for SW

```python
sns.boxplot(x='Classification', y='SW', data = iris)
plt.title('Boxplot for SW', color='green', size = 20)
plt.show()
```

## Boxplot for PL

```python
sns.boxplot(x='Classification', y='PL', data = iris)
plt.title('Boxplot for PL', color='hotpink', size = 20)
plt.show()
```

## Boxplot for PW

```python
sns.boxplot(x='Classification', y='PW', data = iris)
plt.title('Boxplot for PW', color='orange', size = 20)
plt.show()
```

## Boxplot for all classification

```python
sns.boxplot(data = iris)
plt.title('Boxplot for all classification', color='orange', size = 20)
plt.show()
```

# Visualization 2:                                           HISTOGRAM

## Histogram for SL using plt.hist()

```python
plt.hist(iris['SL'], color = 'y', rwidth=.9, cumulative=False)   
plt.title('Histogram for SL', size = 21, color = 'purple')
# plt.legend('Classification')
plt.xlabel('SL',size = 15, color = 'b')
plt.ylabel('Count',size = 15, color = 'b')
plt.show()
```

## Histogram for SL using sns.hist()

```python
sns.histplot(x = 'SL', data = iris , hue = 'Classification')   
plt.title('Histogram for SL', size=20, color='indigo')
plt.show()
```
