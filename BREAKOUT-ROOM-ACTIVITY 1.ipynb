# Bar Plots

```python
plt.subplot(2, 3, 1)
sns.barplot(x=df.sex, y=df.total_bill)
plt.title("SEX - TOTAL BILL", color='g',size= 12)
plt.xlabel("sex", color = 'blue', size = 10)
plt.ylabel("total bill",  color = 'blue', size = 10)
plt.subplots_adjust(left=1, bottom=1, right=2, top=3, wspace=1, hspace=1)
                 

plt.subplot(2,3,2)
sns.barplot(x=df.day ,y=df.total_bill)
plt.title("DAY - TOTAL BILL", color="g", size = 12)
plt.xlabel("day",color="b",size = 10)
plt.ylabel("total bill",color="b", size = 10)



plt.subplot(2, 3, 3)
sns.barplot(x=df.time, y=df.total_bill)
plt.title("TIME - TOTAL BILL", color="g",size = 12)
plt.xlabel("time",size = 10,color="b")
plt.ylabel("total bill",size = 10,color="b")


plt.subplot(2,3,4)
sns.barplot(x=df.sex, y=df.tip) 
plt.title("SEX - TIP", color='g',size= 12)
plt.ylabel("tip", color = 'blue', size = 10)
plt.xlabel("sex",  color = 'blue', size = 10)


plt.subplot(2,3,5)
sns.barplot(x=df.day, y=df.tip) 
plt.title("DAY - TIP", size = 12,color="g")
plt.xlabel("day", color = 'b', size = 10)
plt.ylabel("tip", color = 'b', size = 10)



plt.subplot(2, 3, 6)
sns.barplot(x=df.time, y=df.tip)
plt.title("TIME - TIP", size = 12,color="g")
plt.xlabel("time", size = 10,color="b")
plt.ylabel("tip", size = 10,color="b")



plt.show()
```

# Box plots

```python
plt.subplot(3,2,1)
sns.boxplot(x='sex', y='total_bill', data = df)
plt.title('Boxplot for sex and total bill', color='indigo', size = 15)
plt.subplots_adjust(left=1, bottom=1, right=2, top=3, wspace=1, hspace=1)


plt.subplot(3,2,2)
sns.boxplot(x='sex', y='tip', data = df)
plt.title('Boxplot for sex and tip', color='indigo', size = 15)


plt.subplot(3,2,3)
sns.boxplot(x='day', y='total_bill', data = df)
plt.title('Boxplot for day and total bill', color='indigo', size = 15)
plt.subplots_adjust(left=1, bottom=1, right=2, top=3, wspace=1, hspace=1)


plt.subplot(3,2,4)
sns.boxplot(x='day', y='tip', data = df)
plt.title('Boxplot for day and tip', color='indigo', size = 15)


plt.subplot(3,2,5)
sns.boxplot(x='time', y='total_bill', data = df)
plt.title('Boxplot for time and total bill', color='indigo', size = 15)
plt.subplots_adjust(left=1, bottom=1, right=2, top=3, wspace=1, hspace=1)


plt.subplot(3,2,6)
sns.boxplot(x='time', y='tip', data = df)
plt.title('Boxplot for time and tip', color='indigo', size = 15)
sns.set_style('darkgrid')
plt.show()
```

# Inference:  
##### 1. Males have a slightly higher total bill compared to females.
##### 2. Total bill is highest on Sundays and lowest on Thursdays.
##### 3. Total bill is higher during dinner compared to lunch.
##### 4. males generally give a higher tip compared to females.
##### 5. tip amount is slightly higher on weekends (Saturday and Sunday) compared to weekdays.
##### 6. tip amount is slightly higher during dinner compared to lunch.

```python
sns.set_style('whitegrid')
sns.catplot(x='day', y='total_bill', data=df, height=3.5, aspect=2, kind='point', palette='turbo',  hue='smoker')
plt.title('Daywise Total Bill of Smoker and Non-Smoker')
plt.show()
```

# Inference:
##### Highest average total bill amount for smokers are on Sunday, followed by Saturday, Thursday and Friday.
##### Highest average total bill amount for non-smokers are on Sunday, followed by Saturday, Friday and Thursday.
##### Weekends are busier and customers may be more likely to spend more money on their meals.

# Histogram Plot

```python
plt.figure(figsize=(9,5))

sns.histplot(x='total_bill',data=df, hue='sex', kde=True)

plt.title('HISTOGRAM OF TOTAL BILL',size = 20,color ='r')
plt.xlabel('TOTAL BILL')
plt.ylabel('COUNT')
plt.show()
```

# Inference: 
##### The distributions are similar in shape for both male and female, with a peak around the 10-20 range. 
##### However, the male distribution is slightly shifted to the right, indicating that male customers tend to have slightly higher total bills on average.
##### The KDE plot overlaid on top of the histograms provides a smooth estimate of the distribution shape for each group. 

```python
sns.pairplot(data=df, hue='day')
plt.suptitle("Daywise Pairplot", y=1.02, size=20, color='green')
plt.show()
```

```python
sns.pairplot(data=df, hue='sex')
plt.suptitle("Sexwise Pairplot", y=1.02, size=20, color='green')
plt.show()
```

# Line plot

```python
plt.figure(figsize=(14, 5))

sns.lineplot(data = df, y="total_bill",  x="day",hue = "sex",palette = "muted")
plt.title("Line plot", size=20, color='indigo')
plt.show()
```

# Strip Plot

```python
# plt.figure(figsize=(14, 5))
sns.stripplot(data=df,  x="day", y="tip",   hue="sex",  dodge=True,  order = [ "Thur", "Fri", "Sat", "Sun"], palette = "muted")
plt.title("Strip plot", size=20, color='indigo')
plt.show()
```

# Swarm Plot

```python
# plt.figure(figsize=(14, 5))
sns.swarmplot(data=df, x="day",  y="tip", hue="sex", dodge=True,  order = [ "Thur", "Fri", "Sat", "Sun"],  palette = "muted")
plt.title("Swarm plot", size=20, color='indigo')
plt.show()
```

# Violin plot

```python
# plt.figure(figsize=(14, 5))
sns.violinplot(data=df, x="day", y="total_bill", hue="sex", dodge=True, order = [ "Thur", "Fri", "Sat", "Sun"],palette = "bright")
plt.title("Violin plot", size=20, color='indigo')
plt.show()
```

# Joint plot

```python
plt.figure(figsize=(14,6))
sns.jointplot(data=df, x="total_bill", y="tip",  hue = 'day')
plt.suptitle("Join plot", y=1.02, size=20, color='indigo') 
plt.show()
```
